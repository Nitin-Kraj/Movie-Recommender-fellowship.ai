# Movie Recommendation

In this notebook I have developed a movie recommender system based on the MovieLens dataset from http://grouplens.org/datasets/movielens/1m/. The files I have used are ratings.dat, movies.dat, users.dat

 Importing the libraries

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import sys
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling
%matplotlib inline
```

 Getting the data from ratings.dat, movies.dat and users.dat files

```python
path = '/Users/nitin/Desktop/Movie recommendation/ml-1m/'
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(path + 'ratings.dat', sep = '::', header = None, names = rnames, engine='python')
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table(path + 'movies.dat', sep = '::', header = None, names = mnames, engine='python')
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table(path + 'users.dat', sep = '::', header = None, names = unames, engine='python')
```

```python
movies.head()
```

```python
users.head()
```

```python
ratings.head()
```

Merging the ratings and users with movies


```python
df = pd.merge(pd.merge(ratings, users), movies)
df.head()
```

```python
pandas_profiling.ProfileReport(df)
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

# Movie recommendation

```python
mean_ratings = df.pivot_table(index = ['user_id'], columns = ['title'], values = 'rating', aggfunc = 'mean')
mean_ratings.head()
```

```python
df
```

```python
starwars_user_ratings = mean_ratings["Star Wars: Episode IV - A New Hope (1977)"]
starwars_user_ratings.head()
```

```python
similar_to_starwars = mean_ratings.corrwith(starwars_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
```
